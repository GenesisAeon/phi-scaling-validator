## EML Tree Depths
Test whether GenesisAeon EML operator tree depths are Φ^(1/3)-spaced.

```python
import sys; sys.path.insert(0, '../src')
from phi_scaling.eml_scaling import EMLScalingAnalyzer, EML_TREE_DEPTHS
from phi_scaling.phi_constants import PHI_CUBEROOT

eml = EMLScalingAnalyzer()
summ = eml.summary()

print("EML tree depth expressions:")
for e in sorted(EML_TREE_DEPTHS, key=lambda x: x.tree_depth):
    print(f"  depth={e.tree_depth}  {e.name}  (P{e.package})")

print(f"\nConsecutive ratios: {summ['ratios']}")
print(f"Mean ratio:  {summ['mean_ratio']:.5f}  (Φ^(1/3) = {PHI_CUBEROOT:.5f})")
print(f"p-value:     {summ['p_value']:.4f}")
print(f"confirmed:   {summ['confirmed']}")
print(f"note:        {summ['note']}")
```
