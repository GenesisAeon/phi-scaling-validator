## Φ-power ladder
All integer and third-integer powers of Φ from 0 to 4.

```python
import matplotlib.pyplot as plt
import numpy as np

powers = np.arange(0, 4.1, 1/3)
values = PHI ** powers

fig, ax = plt.subplots(figsize=(10, 4))
ax.semilogy(powers, values, 'o-', color='goldenrod', linewidth=2, markersize=8)
ax.axhline(PHI_CUBEROOT, color='red', linestyle='--', alpha=0.5, label=f'Φ^(1/3) = {PHI_CUBEROOT:.4f}')
ax.set_xlabel('Exponent n')
ax.set_ylabel('Φ^n')
ax.set_title('Golden Ratio Power Ladder — GenesisAeon P38')
ax.legend(); ax.grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
print("Φ^(1/3) step confirmed as fundamental inter-scale spacing")
```
